# Analyze excluded concepts

```python
import glob 
import os 
from collections import defaultdict
from collections import Counter

from utils import read_group, read_csv, to_csv
```

```python
exp_group = 'experiment3'
prop = 'used_in_cooking'
collection = 'complex'

path_filtered = f'../data_pair_filtering/aggregated/{exp_group}/{prop}.csv'
print(path_filtered)
data_annotated = read_csv(path_filtered)

path_all = f'../full_data/concepts_additional_info_ids_updated/{collection}/{prop}.csv'
all_candidates = read_csv(path_all)

all_candidates_dict = dict()
for d in all_candidates:
    concept = d['lemma'].strip()
    all_candidates_dict[concept] = d


cat_cnt = Counter()
label_cnt = Counter()

concepts_include = [d['lemma'] for d in data_annotated if d['decision'] == 'include']

for concept in concepts_include:
    info_dict = all_candidates_dict[concept.strip()]
    cat = info_dict['categories_str']
    label = info_dict['label']
    cat_cnt[cat] += 1
    label_cnt[label] +=1
    
print('categories')
print()
for cat, cnt in cat_cnt.most_common():
    print(cat, cnt)
print('\n-----\n')
print('labels')
print()
for l, cnt in label_cnt.most_common():
    print(l, cnt)
```

```python
# concepts not in dataset (not selected in sampling)
cat_cnt_all = Counter()
for d in all_candidates:
    concept = d['lemma']
    cat = d['categories_str']
    if concept not in concepts_include:
        cat_cnt_all[cat] += 1
        print(concept, cat)
    
```

```python
for cat, cnt in cat_cnt_all.most_common():
    print(cat, cnt)
```
